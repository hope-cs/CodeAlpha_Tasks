```python
from IPython.display import clear_output
```

```python
RESPONSES = {
    ("hello", "hi", "hey", "good morning", "good evening"): 
        "👋 Hello! How can I help you today?",
    
    ("how are you", "how are you doing", "how do you do"): 
        "😊 I'm doing great, thanks for asking! What about you?",
    
    ("what is your name", "who are you", "your name", "name"): 
        "🤖 I'm PyBot, your Python-powered assistant!",
    
    ("what can you do", "help", "commands"): 
        "💡 Try saying: hello, how are you, your name, joke, time, bye",
    
    ("joke", "tell me a joke", "say something funny"): 
        "😄 Why do programmers prefer dark mode? Because light attracts bugs! 🐛",
    
    ("bye", "goodbye", "exit", "quit", "see you"): 
        "👋 Goodbye! Have a great day!",
    
    ("thanks", "thank you", "thx"): 
        "🙏 You're welcome! Anything else I can help with?",
    
    ("time", "what time is it", "current time"): 
        "DYNAMIC_TIME",
}

print("Responses are successfully added.")
```

```python
import datetime

def get_response(user_input):
    text = user_input.lower().strip()
    
    for keywords, response in RESPONSES.items():
        if any(keyword in text for keyword in keywords):
            if response == "DYNAMIC_TIME":
                now = datetime.datetime.now().strftime("%I:%M %p")
                return f"🕐 Current time is {now}"
            return response
    
    return "🤔 I didn't understand that. Type 'help' to see what I can do!"

print("Response function works fine.")
```

```python
def display_chat(history):
    print("=" * 45)
    print("         🤖 PyBot — Python Chatbot")
    print("      Type 'bye' to exit the chat")
    print("=" * 45)
    
    for sender, message in history:
        if sender == "You":
            print(f"\n  🧑 You:   {message}")
        else:
            print(f"  🤖 PyBot: {message}")
    print()
```

```python
def chatbot():
    history = []
    
    while True:
        clear_output(wait=True)
        display_chat(history)
        
        user_input = input("You: ").strip()
        
        if not user_input:
            continue
        
        response = get_response(user_input)
        
        history.append(("You", user_input))
        history.append(("PyBot", response))
        
        if any(word in user_input.lower() for word in ("bye", "goodbye", "exit", "quit")):
            clear_output(wait=True)
            display_chat(history)
            break

chatbot()
```
