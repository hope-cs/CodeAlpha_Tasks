# 🎮 Hangman Game
**CodeAlpha Python Internship — Task 1**

A classic text-based word guessing game built in Python.
The player guesses a hidden word one letter at a time, with a maximum of 6 wrong guesses.

## 📦 Step 1: Import Libraries

```python
import random
from IPython.display import clear_output
```

## 🖼️ Step 2: Define Hangman Drawing Stages
Each stage represents a body part added after a wrong guess.
Index 0 = no wrong guesses, Index 6 = full hangman (game over).

```python
HANGMAN_STAGES = [
    """
       -----
       |   |
           |
           |
           |
           |
    =========""",
    """
       -----
       |   |
       O   |
           |
           |
           |
    =========""",
    """
       -----
       |   |
       O   |
       |   |
           |
           |
    =========""",
    """
       -----
       |   |
       O   |
      /|   |
           |
           |
    =========""",
    """
       -----
       |   |
       O   |
      /|\\  |
           |
           |
    =========""",
    """
       -----
       |   |
       O   |
      /|\\  |
      /    |
           |
    =========""",
    """
       -----
       |   |
       O   |
      /|\\  |
      / \\  |
           |
    ========="""
]
```

## 📝 Step 3: Define Word List
The game randomly picks one word from this list each time it starts.

```python
WORDS = ["python", "developer", "hangman", "programming", "keyboard"]
```

## 🔡 Step 4: Helper Function — Display Word Progress
Shows correctly guessed letters and hides the rest with underscores.

```python
def display_word(word, guessed_letters):
    """
    Returns the word with guessed letters revealed and unguessed letters as '_'.
    Example: word='python', guessed={'p','n'} → 'p _ _ _ _ n'
    """
    return " ".join(letter if letter in guessed_letters else "_" for letter in word)
```

## 🎮 Step 5: Main Game Function
Controls the full game loop — input, validation, win/loss conditions.

```python
def hangman():
    """
    Main Hangman game function.
    - Picks a random word
    - Accepts letter guesses from the player
    - Updates and redraws the screen after each guess
    - Ends on win (all letters guessed) or loss (6 wrong guesses)
    """
    word = random.choice(WORDS)       # Pick a random word
    guessed_letters = set()           # Set stores unique guessed letters
    wrong_guesses = 0                 # Counter for wrong guesses
    max_wrong = 6                     # Max allowed wrong guesses
    message = ""                      # Feedback message from last guess

    while wrong_guesses < max_wrong:

        clear_output(wait=True)       # Clear previous output to avoid long scroll

        # --- Redraw entire game state ---
        print("=" * 40)
        print("       🎮 Hangman Game")
        print("=" * 40)
        print(HANGMAN_STAGES[wrong_guesses])
        print(f"\nWord:  {display_word(word, guessed_letters)}")
        print(f"Wrong guesses left: {max_wrong - wrong_guesses}")
        print(f"Letters guessed: {', '.join(sorted(guessed_letters)) if guessed_letters else 'None yet'}")

        # Show result of previous guess
        if message:
            print(f"\n{message}")

        # --- Check win condition ---
        if all(letter in guessed_letters for letter in word):
            print(f"\n✅ YOU WIN! The word was: '{word.upper()}'")
            print("🎉 Congratulations!")
            return

        # --- Get player input ---
        guess = input("\nGuess a letter: ").lower().strip()

        # Validate: single alphabetic character only
        if len(guess) != 1 or not guess.isalpha():
            message = "⚠️  Please enter a single letter only!"
            continue

        # Validate: no duplicate guesses
        if guess in guessed_letters:
            message = f"⚠️  You already guessed '{guess}'. Try another!"
            continue

        guessed_letters.add(guess)

        # --- Check if guess is correct ---
        if guess in word:
            message = f"✅ Nice! '{guess}' is in the word!"
        else:
            wrong_guesses += 1
            message = f"❌ '{guess}' is NOT in the word."

    # --- Game Over ---
    clear_output(wait=True)
    print("=" * 40)
    print("       🎮 Hangman Game")
    print("=" * 40)
    print(HANGMAN_STAGES[wrong_guesses])
    print(f"\nWord:  {display_word(word, guessed_letters)}")
    print(f"\n💀 GAME OVER! The word was: '{word.upper()}'")
    print("Better luck next time!")
```

## ▶️ Step 6: Run the Game
Execute the cell below to start playing!

```python
hangman()
```
