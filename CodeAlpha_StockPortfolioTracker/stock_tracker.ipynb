```python
import csv
import os
```

```python
STOCK_PRICES = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOGL": 140,
    "AMZN": 185,
    "MSFT": 420
}
```

```python
def display_available_stocks():
    print("\n📈 Available Stocks:")
    print(f"{'Symbol':<10} {'Price (USD)':>12}")
    print("-" * 24)
    for symbol, price in STOCK_PRICES.items():
        price_str = f"${price}"
        print(f"{symbol:<10} {price_str:>12}")

display_available_stocks()
```

```python
def get_portfolio():
    portfolio = {}
    print("\n💼 Enter your stock holdings (type 'done' to finish):")

    while True:
        symbol = input("Stock symbol: ").upper().strip()

        if symbol == "DONE":
            break

        if symbol not in STOCK_PRICES:
            print(f"⚠️  '{symbol}' not found. Available: {', '.join(STOCK_PRICES)}")
            continue

        try:
            qty = int(input(f"Quantity of {symbol}: "))
            if qty <= 0:
                print("⚠️  Quantity must be a positive number.")
                continue
        except ValueError:
            print("⚠️  Please enter a valid number.")
            continue

        portfolio[symbol] = qty

        # Print confirmation after each entry
        value = STOCK_PRICES[symbol] * qty
        print(f"✅ Added {qty} shares of {symbol} @ ${STOCK_PRICES[symbol]} = ${value:,}")
        print("-" * 40)

    return portfolio
```

```python
def display_summary(portfolio):
    if not portfolio:
        print("⚠️  No stocks entered.")
        return None, None

    print("\n📊 Portfolio Summary:")
    print(f"{'Symbol':<10} {'Qty':>6} {'Price':>10} {'Value':>12}")
    print("-" * 42)

    total = 0
    rows = []

    for symbol, qty in portfolio.items():
        price = STOCK_PRICES[symbol]
        value = price * qty
        total += value
        rows.append((symbol, qty, price, value))
        print(f"{symbol:<10} {qty:>6} ${price:>9} ${value:>11,}")

    print("-" * 42)
    print(f"{'TOTAL INVESTMENT':>28}  ${total:>11,}")

    return rows, total
```

```python
def save_to_file(rows, total):
    choice = input("\n💾 Save portfolio? (csv/txt/no): ").lower().strip()

    if choice == "csv":
        with open("portfolio.csv", "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["Symbol", "Quantity", "Price (USD)", "Value (USD)"])
            for row in rows:
                writer.writerow(row)
            writer.writerow(["", "", "TOTAL", total])
        print("✅ Saved to portfolio.csv")

    elif choice == "txt":
        with open("portfolio.txt", "w") as f:
            f.write("Stock Portfolio Report\n")
            f.write("=" * 42 + "\n")
            for symbol, qty, price, value in rows:
                f.write(f"{symbol:<10} {qty:>6} ${price:>9} ${value:>11,}\n")
            f.write("-" * 42 + "\n")
            f.write(f"{'TOTAL':>28}  ${total:>11,}\n")
        print("✅ Saved to portfolio.txt")

    else:
        print("📋 Portfolio not saved.")
```

```python
from IPython.display import clear_output

def main():
    message = ""

    while True:
        clear_output(wait=True)
        display_available_stocks()

        if message:
            print(f"\n{message}")

        portfolio = get_portfolio()
        clear_output(wait=True)
        result = display_summary(portfolio)

        if result[0] is not None:
            rows, total = result
            save_to_file(rows, total)

        again = input("\n🔄 Track another portfolio? (yes/no): ").lower().strip()
        if again != "yes":
            print("\n👋 Goodbye!")
            break

main()
```
